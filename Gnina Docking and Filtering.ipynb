```python
from rdkit import Chem
from rdkit.Chem import AllChem
```

```python
import pandas as pd
```

```python
  gnina_poses= Chem.SDMolSupplier('/Users/sybrenvandenbedem/Documents/Stanford Internship/Smina/second100k_out_best_gnina.sdf')
```

```python
mols = [p for p in gnina_poses if float(p.GetProp('CNNaffinity')) > 5.1]
w = Chem.SDWriter('second100k_out_best_gnina_above5_1.sdf')
for m in mols: w.write(m)
```

```python
#make a table
mols = [p for p in gnina_poses if float(p.GetProp('CNNaffinity')) > 5.1]
tbl = []
for m in mols:
    tbl.append([m.GetProp('_Name'), m.GetProp('CNNaffinity')])
df = pd.DataFrame(tbl)
df.to_csv('/Users/sybrenvandenbedem/Documents/Stanford Internship/Smina/second100k_out_best_gnina_table')
df.head(60)
```
